# Sklearn

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
n = 30
x = np.arange(30)
```

```python
x
```

```python
y = 4 *x - 7 + (5*np.random.randn(n))
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
#requires 2D
model.fit(x[:,np.newaxis],y)
yfit = model.predict(x[:,np.newaxis])
plt.scatter(x,y)
plt.plot(x,yfit,color='r')
```

# Support Vector Machine

```python
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples = 500, centers = 3, cluster_std = 1.2)
plt.scatter(x[:,0],x[:,1],c=y,cmap='autumn')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33)
```

```python
from sklearn.svm import SVC
```

```python
mdl = SVC(gamma='auto')
mdl.fit(x_train,y_train)
```

```python
y_predict = mdl.predict(x_test)
```

```python
acc = 100*np.sum(y_predict == y_test)/y_test.size
print(f'accuracy is {acc}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
mdl = RandomForestClassifier(n_estimators = 20)
mdl.fit(x_train,y_train)
```

```python
y_predict = mdl.predict(x_test)
ac = 100*np.sum(y_predict)/y_test.size
print(ac)
```
